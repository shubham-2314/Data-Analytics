```python
feature=["Target"]
y =pd.DataFrame(data=calihouses.target,columns=feature)
df_final=pd.concat([df,y],axis=1)
```

```python
from scipy.stats import zscore
z=np.abs(zscore(df_final))
z
```

```python
df_final=df_final[(z<3).all(axis=1)]
```

```python
x=df_final.iloc[:,:-1].values
x.shape
```

```python
y = df_final.iloc[:,-1].values
y=y.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.22,random_state=50)
```

```python
from sklearn.metrics import mean_squared_error
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
lm.coef_
```

```python
lm.intercept_
```

```python
lm.score(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error
pred=lm.predict(x_test)
print("predicted result Target",pred)
print("actual Target",y_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,pred))
```

```python
def linreg():
    lm.fit(x,y)
    print(lm.score(x,y))
    print(r2_score(y_test,pred))
    print(mean_squared_error(y_test,pred))
```
