```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score as score
import warnings
warnings.filterwarnings("ignore")
```

```python
lc=pd.read_csv("lung cancer dataset.csv",header=0)
```

```python
lc
```

```python
df=pd.DataFrame(data=lc)
```

```python
df.columns = range(df.shape[1])
```

```python
df.isnull().sum()
```

```python
df.corr()
```

```python
df.iloc[:,4]=pd.to_numeric(df.iloc[:,4], errors='coerce')
df.iloc[:,38]=pd.to_numeric(df.iloc[:,38], errors='coerce')
print(df.iloc[:,4])
print(df.iloc[:,38])
```

```python
df=df.fillna(0)
print(df.iloc[:,4])
print(df.iloc[:,38])
```

```python
x=df.iloc[:,1:-1].values
y=df.iloc[:,0].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.22,random_state=50)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
def rfc():
    rfc = RandomForestClassifier(n_estimators=100, n_jobs=-1, )
    rfc.fit(x_train, y_train)
    pred = rfc.predict(x_test)
    print('Accuracy in percent = ',score(pred, y_test)*100)
    print(f1_score(y_test, pred, average='macro'))
    print(f1_score(y_test, pred, average='micro'))
    print(f1_score(y_test, pred, average='weighted'))
```

```python
rfc()
```

```python
def lr():
    lr = LogisticRegression()
    lr.fit(x_train, y_train)
    pred = lr.predict(x_test)
    print('Accuracy in percent = ',score(pred, y_test)*100)
    print(f1_score(y_test, pred, average='macro'))
    print(f1_score(y_test, pred, average='micro'))
    print(f1_score(y_test, pred, average='weighted'))
    
```

```python
lr()
```
