```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings("ignore")
```

```python
features=["buying","maint","doors","persons","lug_boot","safety","Class Values"]
car=pd.read_csv("car.csv",names=features,header=None)
```

```python
car.head()
```

```python
df=pd.DataFrame(data=car)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df_new=pd.get_dummies(df,columns=['buying','maint','doors','persons','lug_boot','safety'])
```

```python
df_new.head()
```

```python
le=LabelEncoder()
df_new["Class Values"]=le.fit_transform(df_new["Class Values"])
```

```python
df_new.head()
```

```python
df_new.dtypes
```

```python
x=df_new.drop(["Class Values"],axis=1).values
```

```python
y=df_new["Class Values"].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.18,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,f1_score
from sklearn.metrics import accuracy_score as score
```

```python
rfc = RandomForestClassifier(n_estimators=100, n_jobs=-1, )
rfc.fit(x_train, y_train)
pred = rfc.predict(x_test)
print('Accuracy in percent = ',score(pred, y_test)*100)
print(f1_score(y_test, pred, average='macro'))
print(f1_score(y_test, pred, average='micro'))
print(f1_score(y_test, pred, average='weighted'))
```
