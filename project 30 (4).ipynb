```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
crime=pd.read_csv("20_Victims_of_rape.csv")
```

```python
df=pd.DataFrame(data=crime)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.Area_Name.value_counts()
```

```python
df.groupby(['Year','Subgroup']).Rape_Cases_Reported.sum()
```

```python
plt.figure(figsize=(8,8))
df.groupby(['Year','Subgroup']).Rape_Cases_Reported.sum().plot(kind='bar');
```

```python
df.Rape_Cases_Reported.plot(kind='hist',bins=20);
```

```python
victims_rape_2010_total = df[(df['Year']==2010) & (df['Subgroup']== 'Total Rape Victims')]
ax1 = victims_rape_2010_total['Victims_of_Rape_Total'].plot(kind='barh',figsize=(20, 15))
ax1.set_xlabel("Number of rape victims (2010)", fontsize=25)
ax1.set_yticklabels(victims_rape_2010_total['Area_Name']);
```

```python
import seaborn as sns
plt.figure(figsize=(8,8))
df_corr=df.corr()
sns.heatmap(df_corr, xticklabels = df_corr.columns.values, yticklabels = df_corr.columns.values,annot=True);
```
